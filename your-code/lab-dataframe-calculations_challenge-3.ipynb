```python
# your code here


#SUB PROBLEM 1: How to aggregate the amount_spent for unique customers?
#We'll aggregate using the group by method and the sum function
customers=orders.groupby(by='CustomerID').sum()
customers
```

```python
#SUB PROBLEM 2: How to select customers whose aggregated amount_spent is in a given quantile range?
#In this case the approach is to bin the data. There are two functions for this: cut and qcut.
#Qcut seems the best one for this case since we don't know the histogram / have a view of how the data is distributed etc.

binning=pd.qcut(customers['amount_spent'], 3)
```

```python
#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

customer_profiles=['avg','preferred','vip']  #first we create a list with the labels for the customers

binning=pd.qcut(customers['amount_spent'], 3, labels=customer_profiles) #we then add the labels to the bins
customers['bins'] = binning
customers
```
